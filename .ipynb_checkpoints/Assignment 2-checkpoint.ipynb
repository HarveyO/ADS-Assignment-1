```python
# import necessary packages 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# read data using pandas in a dataframe

df = pd.read_csv("ADS-Assignment-2-3/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
# check information of the dataset

df.info()
```

```python
# display the columns in the data
df.columns
```

```python
# check the number of rows and columns in the dataset

df.shape
```

```python
# filter data by selecting meanful columns

df_new = df[["Age","Attrition","BusinessTravel","Department","DistanceFromHome","EducationField","Gender",
            "JobRole","MonthlyIncome","HourlyRate","MaritalStatus","YearsAtCompany"]]
df_new.info()
```

```python
# check the new data frame
df_new.head()
```

```python
# showing the correlation between each variable
sns.pairplot(df_new, hue="Attrition", markers=["o","s"], diag_kind='kde', plot_kws={'alpha': 0.6, 's': 80, 'edgecolor':'k'}, height = 4)
plt.title("Relationships between the variables")
plt.show()
```

Using a pairplot to show relationship between the various columns in the dataset, it shows that there are no correlations between the various variables meanwhile just a weak correlation between Age and YearsAtCompany unlike others with zero relationships.

```python
# group & count the total number of attrition in the dataset according to variation

attri_distr = df_new["Attrition"].value_counts().reset_index()
attri_distr
```

Without any measure of calculation, it could be described that from the dataset, attrition on the members of the company had lower effect than otherwise. 

```python
# plot showing the distribution of the number of attrition according to variation
sns.barplot(x="index", y="Attrition", data=attri_distr)
plt.show()
```

```python
# check for the minimum & maximum age of worker in the company

df_new.agg({"Age":["min","max"]})
```

Also observed that the youngest person affected in the company is 18 years of age while the oldest is about 60 years respectively.

```python
# group the number of attrition by marital status in the dataset

data_marital_attri = df.groupby("MaritalStatus")["Attrition"].count().reset_index()
x =pd.DataFrame(data_marital_attri)
x
```

```python
# plot showing the distribution of attrition on marital status

sns.barplot(x="MaritalStatus", y="Attrition", data=x)
plt.show()
```

Considering the effect of attrition, it was revealed that the married people working in this company are most likely more affected. Maybe because married people have more responsibilities on them unlike the divorced group who tends to be less affected.

```python
# group & count number of attrition by gender

gender = df.groupby("Gender")["Attrition"].count().reset_index()
gender_df =pd.DataFrame(gender)
gender_df
```

```python
# plot showing the distribution of attrition on gender

sns.barplot(x="Gender", y="Attrition", data=gender_df)
plt.show()
```

It can be assumed from the above figure that male workers in the company are more affected probably because most of the tasks are assigned to male than female. 

```python
# count number of attrition by department

department = df.groupby("Department")["Attrition"].count().reset_index()
department_df = pd.DataFrame(department)
department_df
```

```python
# plot showing the distribution of attrition according to various department

sns.barplot(x="Department", y="Attrition", data=department_df)
plt.show()
```

No doubt that the research department does more work compared to other departments of the company as shown in the graph above. Maybe they spend more time in researching and developing a structure to implement a better standard for the company or possibly there are more employees in Research & Development department unlike the others.

```python
# count number of attrition by business travel

business = df.groupby("BusinessTravel")["Attrition"].count().reset_index()
business_df = pd.DataFrame(business)
business_df
```

```python
# plot showing distribution of attrition according to business travel

sns.barplot(x="BusinessTravel", y="Attrition", data=business_df)
plt.show()
```

Furthermore, we discovered that members who travel rarely tend to be affected more compared to non-travelers and frequent travelers.

```python
# count number of attrition by job role

jobrole = df.groupby("JobRole")["Attrition"].count().reset_index()
jobrole_df = pd.DataFrame(jobrole)
jobrole_df.head()
```

```python
# plot showing distribution according to job roles in the company
sns.barplot(x="JobRole", y="Attrition", data=jobrole)
plt.xticks(rotation=90)
plt.show()
```

At a close scope, we dug into details a bit further to discover the job roles with intensive pressure and found out that Sales Executives are majorly affected from the figure above. Next to Research Scientist followed by Laboratory Technicians, Manufacturing Directors, Healthcare Representatives, Managers, Sales Representatives, Research Directors and Human Resources.

```python
# distribution of monthly income in various departments, showing the age and gender as well as level of attrition
fig = px.bar(df_new, x='Department', y='MonthlyIncome',
             hover_data=['Gender', 'Age'], 
             color='Attrition',  
             height=500)
# show data 
fig.show()
```

The chart above uncovers the monthly income of the members of the company according to the various department. Moreso, showing the variability in age, gender and level of attrition. In that regard, Research & Development department has a high monthly income and attrition. Though as shown with low positive attrition compared to negative attrition found in the department. It also describes the age categories in the various departments, that there are mostly aged people above 40 in the department. This could be as a result of high experience in order to deliver on the job. Unlike Research & Development department, Sales department has more younger people below the age of 45. This must be as a result of strong marketing power to pull sales trend to the opposite direction of negativity. It led to the display of lower positive attrition compared to the negative attrition found in the department. Human Resources department has shown scanty effect of attrition as most of them in the department are older with more experience on their jobs. Therefore revealing less pressure to deliver on the job.

```python
# breakdown of the effect of attrition by jobrole and distance from home 
data_marital_attr = df.groupby(["Attrition","JobRole"])["DistanceFromHome"].count().reset_index()
y =pd.DataFrame(data_marital_attr)
y.head()
```

```python
# plot showing the effect of attrition by jobrole and distance from home
sns.barplot(x="JobRole", y="DistanceFromHome", hue="Attrition", data=y)
plt.xticks(rotation=90)
plt.show()
```

We also observed that Laboratory Technicians have the highest number of positive attrition compared to others though with a less negative attrition at close range with Sales Executives. Next is the Sales Executives followed by Research Scientists and Sales Representatives. At this same comparison, Research Directors have the most minimal positive attrition followed by Managers then Manufacturing Director, Human Resources and Healthcare Representatives.

```python
# breakdown of the effect of attrition by education level and average monthly income
monthly_income_mean =df.groupby(["Attrition","Education"])["MonthlyIncome"].mean().reset_index()
df_MIM = pd.DataFrame(monthly_income_mean)
df_MIM
```

```python
# plot showing the effect of attrition by education level and average monthly income
sns.barplot(x="Education", y="MonthlyIncome", hue="Attrition", data=df_MIM)
plt.show()
```

The above chart reveals a growth in salary as level of education increases which also describes level of attrition along side. Which may furtherly describe that, the higher the educational level, the more work loads on members of the company/organization.
